# Plotting Graphs

### Getting Started

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.scatter(x,y)
# To bring the line
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

# First char will decide the color, how do we want our line to show up
# The following patterns are available "o,-,^,p,--,-.,+" etc.
plt.plot(x2,y2,'b--')
plt.plot(x,y,'ro')
plt.show()
```

```python
# Plot the funciton x^3
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0,5, 0.1)
y = x**3
plt.plot(x,y)
plt.show()
```

### Customising Graphs

```python
x = np.arange(0,5, 0.1)
y = x**3
y2 = x**2
# plt.plot(x,y, color = 'black', marker = '+')
plt.plot(x,y, color = 'black', linewidth=5, label = 'x^3')
plt.plot(x,y2, color = 'red', linewidth=5, label = 'x^2')

# The axis size - Passing first the x label and the y label
plt.axis([0,10,0,150])

plt.ylabel('y')
plt.xlabel('x')

plt.title('Matplotlib demo')
# Shows text in the given coordinates
plt.text(2,80,'Text', fontsize = '15')
# to show the labels
plt.legend()

plt.grid()
plt.show()
```

### Bubble Charts

```python
import numpy as np
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
c = np.arange(len(year))
c
```

```python
# S is the size of the bubbles, c is the color alpha is the opacity edgecolors five the border color of bubles
# In case c contains some values of year to be same then thei color will be the same
plt.scatter(year, salary, s= population, c = c, alpha=0.5, edgecolors='black')
plt.title("Bubble Chart")
# To show the number corresponding to the bubbles - We could possibly run a for loop to print the numbers
plt.text(year[0]+ 0.1, salary[0]+0.1, population[0])
plt.show()
```

### Pie Chart

```python
labels = ['A','B','C','D']
sizes = [3, 4, 6, 2]
colors = ['blue', 'purple', 'red', 'pink']
plt.title('Split amoung axis')

# to highlight a particular graph
explode = [0.1,0,0,0]

# in autopct we can wither pass our own function or if we pass %.1f it will give the percentage split 
# bydefaul the graph is clockwise, but we can make it counter clockwise
# We can give the startangle which helps us to decide from where the graph starts
plt.pie(sizes, colors = colors, labels = labels, autopct='%.1f%%', explode=explode, counterclock=False, 
        startangle=100)
# I did not require it but it helps if the graph is streached
plt.axis('equal')
plt.show()
```

### Histogram

```python
a = [1,2,3,4,1,3,4,2,3,1,2]
xt = np.arange(23)
xt
```

```python
plt.hist(a)
plt.show()
```

```python
# By default it creates 10 bins(towers) (max-min)/no. of bins in our case (22-1)/10 = 2.1
b = [1, 2, 1, 3, 5, 4, 5, 5, 4, 3, 3, 10, 7, 6, 7, 8, 8, 20, 22, 11, 14]
# If we want don't want the bins that are created automatically then we have to use 'bins'
# It follows the format [a,b) except in the last bin where it is [a,b]
plt.hist(b, bins = 21, edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

```python
# If we wanted bins to be distributed a particular way then we can do that
plt.hist(b, bins = [1,3,6,10,15,20], edgecolor='black')
plt.axis([0,25,0,10])
plt.show()
```
